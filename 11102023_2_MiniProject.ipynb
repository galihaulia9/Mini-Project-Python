## 1.A

Lakukan running dengan iterasi 10x untuk membaca input dari user yang berupa nama kepala keluarga dan penghasilan nya
i. Contoh setiap baris input: 'Adi alkatiri | 6.7'

Yang menunjukkan bahwa kepala keluarga nya adalah Adi alkatiri dan penghasilan bulanan keluarganya adalah 6.7

ii. Kemudian ditentukan status ekonomi-nya dengan kriteria sebagai berikut->

#Jika penghasilan diatas 15 juta, maka statusnya crazy rich

#Jika penghasilan antara 5 juta sampai 15 juta, maka statusnya makmur

#Jika penghasilan antara 2 juta sampai 5 juta, maka statusnya mandiri

#Jika penghasilan antara 1 juta sampai 2 juta, maka statusnya butuh
pendampingan

#Jika penghasilan antara 0 sampai 1 juta, maka statusnya rentan

```python
# Initialize the economic status list
status_ekonomi = {
    'crazy rich': [],
    'makmur': [],
    'mandiri': [],
    'butuh pendampingan': [],
    'rentan': []
}

# Iterate for 10 times
for i in range(10):
    input_data = input(f"Masukkan data keluarga ke-{i+1} (nama | penghasilan): ")
    
    # Separate name and income using "|" as separator
    data = input_data.split('|')
    
    # To ensure there are two parts to the input data (name and income)
    if len(data) != 2:
        print("Format input salah. Gunakan format 'nama | penghasilan'.")
        continue
    
    nama_keluarga = data[0].strip()
    penghasilan = float(data[1])
    
    if penghasilan > 15000000:
        status = 'crazy rich'
    elif 5000000 <= penghasilan <= 15000000:
        status = 'makmur'
    elif 2000000 <= penghasilan < 5000000:
        status = 'mandiri'
    elif 1000000 <= penghasilan < 2000000:
        status = 'butuh pendampingan'
    else:
        status = 'rentan'
    
    status_ekonomi[status].append(nama_keluarga)

# Display survey results
print("\nHasil survei kondisi sosial ekonomi masyarakat di kampung Durian Runtuh:")
for status, keluarga in status_ekonomi.items():
    print(f"Status '{status}': {', '.join(keluarga)}")
```

EXPLANATION

1. Initialize the economic status according to the provisions of the problem, namely crazy rich, makmur, mandiri, butuh pendampingan and rentan.
2. Perform iterations 10 times by inputting data entering the family name and income in millions. Then separate the results of data input using split (|). Then to ensure the inputted data has two parts, namely name and income using the formula "if len" and if the user enters the wrong input it will issue a sentence "Incorrect input format. Use the format 'name | income'." and the system will repeat the input again. 
3. Next is the determination of economic status with nama_keluarga as data (0) and penghasilan as the 1st data.
4. Display the survey results using the "for" loop function to traverse the "status_ekonomi" dictionary and print the results of user input.

## 2

Anda adalah seorang Dosen di kelas Dasar Pemrograman yang memiliki 40 mahasiswa.

Ada 3 nilai berikut:

np.random.normal(mean, var, n)

tugas ~ (85, 7, 40)

UTS ~ (60, 15, 40)

UAS ~ (75, 10, 40)

Simpan data dalam bentuk Data Frame.

Kemudian buat kolom baru (Nilai Akhir) yang dihitung dengan formula:

tugas: 20%

UTS: 40%

UAS: 40%

Buat kolom baru dan lakukan binning dengan kriteria berikut:

A : 85-100

A- : 80-85

B+ : 75-80

B : 70-75

B- : 65-70

C+ : 60-65

C : 55-60

C- : 50-55

D : 20-50

E : 0-20


```python
import pandas as pd
import numpy as np

# Randomly generate assignment, midterm, and final exam score data
np.random.seed(0)  # For consistent results
nilai_tugas = np.random.normal(85, 7, 40)
nilai_uts = np.random.normal(60, 15, 40)
nilai_uas = np.random.normal(75, 10, 40)

# Initialize a list to store student data
data_mahasiswa = []

# student name input
for i in range(40):
    nama_mahasiswa = input(f"Masukkan nama Mahasiswa {i+1}: ")
    data_mahasiswa.append((nama_mahasiswa, nilai_tugas[i], nilai_uts[i], nilai_uas[i]))

# Creating a DataFrame
data = pd.DataFrame(data_mahasiswa, columns=['Nama Mahasiswa', 'Tugas', 'UTS', 'UAS'])

# Calculating the final score with the given formula
data['Nilai Akhir'] = 0.2 * data['Tugas'] + 0.4 * data['UTS'] + 0.4 * data['UAS']

# Create a new column for binning criteria
bins = [0, 20, 50, 55, 60, 65, 70, 75, 80, 85, 100]
labels = ['E', 'D', 'D', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A']
data['Predikat'] = pd.cut(data['Nilai Akhir'], bins=bins, labels=labels, ordered=False)

```
